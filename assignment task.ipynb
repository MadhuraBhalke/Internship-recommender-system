```python
df.drop(df.columns[[0]], axis=1, inplace=True) 
```

```python
clean = []
import string
for i in df['bio']:
    i = i.lower()
    words = i.translate(str.maketrans('', '', string.punctuation))
    clean.append(words)
```

```python
df['bio'] = clean
```

```python
df.head()
```

```python
from string import digits

cleaned =[]
for l in df['bio']:
    letters = l.translate(str.maketrans('', '', digits))
    cleaned.append(letters)
```

```python
df['bio'] = cleaned
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
sw = stopwords.words('english')
```

```python
clean_sw = [word for word in df['bio'] if word not in sw]
```

```python
df['bio'] = clean_sw
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000,stop_words='english')
```

```python
vector = cv.fit_transform(df['bio']).toarray()
```

```python
vector.shape
```

```python
vector_target = cv.fit_transform(df['job_role']).toarray()
```

```python
vector_target.shape
```

```python
new = pd.DataFrame(vector)
```

```python
new.head()
```

```python
final =  pd.concat([new, df], axis=1)
```

```python
final.head()
```

```python
final.shape
```

```python
final.drop(['bio'],axis=1).head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(job):
    index = final[final['job_role'] == job].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final.iloc[i[0]].job_role)
```

```python
final['job_role'].head()
```

```python
recommend('business development')
```
